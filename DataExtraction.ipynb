```python
from PyPDF2 import PdfReader
import re
import pandas as pd
import numpy as np
```

```python
reader = PdfReader("PDFMINEDEMO.pdf")
```

```python
count = 1
x = 0
page_list = []
for i in range(len(reader.pages)):
    count += 1
    page = reader.pages[i]
    page_list += [page.extract_text().split("\n")]
```

```python
counter = 1
herb_names = []
herb_names_dict = {}
temp = ""
for page in page_list:
    if(page[0][:-1]):
        num = page[0].split(" ")[0][:-1]
        name = page[0].split(" ")[1]
        if(num.isdigit()):
            herb_names_dict[name] = {} 
            temp = name
        else:
            for ele in page:
                if "Rasa :" in ele:
                    herb_names_dict[temp]["Rasa"] = ele[6:]
                if "Guna :" in ele:    
                    herb_names_dict[temp]['Guna'] = ele[6:]
                if "Virya :" in ele:
                    herb_names_dict[temp]['Virya'] = ele[7:]
                if "Vipaka" in ele:
                    herb_names_dict[temp]['Vipaka'] = ele[8:]
                if "IMPORTANT FORMULATIONS " in ele:
                    herb_names_dict[temp]['formulations'] = ele[26:]
                if "THERAPEUTIC USES" in ele:
                    herb_names_dict[temp]['therapeutic_uses'] = ele[20:]
                if "DOSE" in ele:
                    herb_names_dict[temp]['dose'] = ele[8:]
print(herb_names_dict)
```

```python
df = pd.DataFrame.from_dict(herb_names_dict)
df = df.transpose()
df
```
